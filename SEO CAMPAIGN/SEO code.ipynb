```python
import pandas as pd 
import re
```

```python
#Read all datasets
df1 = pd.read_csv(r'c:\Users\DA Admin\Desktop\SEO_CAMPAIGN\Uniace_1.csv')
df2 = pd.read_csv(r'c:\Users\DA Admin\Desktop\SEO_CAMPAIGN\Uniace_2.csv')
df3 = pd.read_csv(r'c:\Users\DA Admin\Desktop\SEO_CAMPAIGN\Uniace_3.csv')
```

```python
#Combine dataFrames and reset the index
df = pd.concat([df1, df2, df3], ignore_index=True)
df
```

```python
df.dtypes
```

```python
#Convert 'DATE' column
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#Extract and add 'hour' column from 'Date' column
df['date'] = df['Date'].dt.date
df['hour'] = df['Date'].dt.hour
```

```python
#Remove unused columns
df = df.drop(['Message Id','Template Id', 'Scenario Id', 'List Id', 'Form Id', 'Campaign Id', 'Campaign Name', 'Scenario Id', 'URL', 'Link', 'Tag'], axis=1, errors='ignore')
```

```python
df
```

```python
#Convert data type
df['MA Referrer'] = df['MA Referrer'].astype(str)

#Compile regex pattern
pattern = re.compile(r'://(?:www\.)?([a-zA-Z0-9\-]+)\.')

#Extract source
df['source'] = df['MA Referrer'].apply(lambda url: re.search(pattern, url).group(1) if re.search(pattern, url) else 'Unknown') 
```

```python
#Convert data type
df['Email']= df['Email'].astype(str)

#Compile regex pattern
df['Domain']=df['Email'].apply(lambda x: x.split('@')[-1])

#Extract source
df['Customer Type'] = df['Domain'].apply(lambda x: 'Người bình thường' if 
   'gmail.com' in x else 
   ('Nhân viên' if 'outlook' in x or 'outlook.com.vn' in x 
   else 'Sinh Viên'))
```
